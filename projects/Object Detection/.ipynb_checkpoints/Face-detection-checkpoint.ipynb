```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
denis =cv2.imread("../../DATA/Denis_Mukwege.jpg")
solvey =cv2.imread("../../DATA/solvay_conference.jpg")
```

```python
plt.imshow(denis,cmap = "gray")
plt.imshow(solvey,cmap="gray")
```

```python
face_cascade = cv2.CascadeClassifier("../../DATA/haarcascades/haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):
    face_img = img.copy()
    
    
    face_rects =face_cascade.detectMultiScale(face_img)
    
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,0),10)
        
    return face_img
    
    
```

```python
result = detect_face(solvey)
plt.imshow(result)
```

```python
def adj_detect_face(img):
    face_img = img.copy()
    
    
    face_rects =face_cascade.detectMultiScale(face_img,scaleFactor = 1.2,minNeighbors=5)
    
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,0),10)
        
    return face_img
    
    
```

```python
result = adj_detect_face(solvey)
plt.imshow(result)
```

```python
eye_cascade = cv2.CascadeClassifier("../../DATA/haarcascades/haarcascade_eye.xml")
```

```python
def detect_eyes(img):
    eye_img = img.copy()
    
    
    eyes_rects =eye_cascade.detectMultiScale(eye_img,scaleFactor = 1.2,minNeighbors=5)
    
    
    for (x,y,w,h) in eyes_rects:
        cv2.rectangle(eye_img,(x,y),(x+w,y+h),(255,0,0),10)
        
    return eye_img
    
    
```

```python
nadia =cv2.imread("../../DATA/Nadia_Murad.jpg")
result = detect_eyes(nadia)
plt.imshow(result,cmap=
          "gray")
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret , frame = cap.read()
    
    frame = detect_face(frame)
    
    cv2.imshow("Video face detect",frame)
    
    if cv2.waitKey(1) & 0xFF == ord("q")
        break
        
        
cap.release()
cv2.destroyAllWindows()
```
