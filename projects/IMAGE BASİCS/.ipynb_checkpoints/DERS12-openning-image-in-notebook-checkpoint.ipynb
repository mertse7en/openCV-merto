```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../DATA/00-puppy.jpg")
type(img)
### hemen typÄ±ne bak 
```

## Resize images

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_rgb.shape
```

```python
img =cv2.resize(img_rgb,(1300,275))
```

```python
plt.imshow(img)
```

## By Ratio oransal


```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
# Bu onemli bak alttabda aÃ§Ä±yo gbi dÃ¼ÅÃ¼n matplotlib qt
%matplotlib qt
plt.imshow(new_img)
```

## Flipping images ****

```python
%matplotlib inline
```

```python
# Along central x axis
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
# Along central y axis
new_img = cv2.flip(new_img,1)
plt.imshow(new_img)
```
